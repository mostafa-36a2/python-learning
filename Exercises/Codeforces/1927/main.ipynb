```python
stdin_generator =""
output_text =""

def stdin_gen(text):
    lines = text.splitlines()
    for line in lines:
        yield line

def set_stdin(text):
    global stdin_generator
    stdin_generator = stdin_gen(text)
    
def input() -> str:
    return next(stdin_generator)


def reset_output():
    global output_text
    output_text = ""

def print(arg):
    global output_text
    output_text+=str(arg)+"\n"

import sys

def assert_case(case_id, text_in,expected_text_out):
    reset_output()
    set_stdin(text_in)
    main()
    if str(expected_text_out).strip(" \n") != str(output_text).strip(" \n"):
        sys.stdout.write(f"XXX Failed at case: {case_id} expected: '{expected_text_out.strip()}'  got :'{output_text.strip()}'\n")
    else:
        sys.stdout.write(f"Succeeded at case: {case_id}\n")
```

```python
def main():
    t = int(input())
    for _ in range(t):
        n, m, k = list(map(int, input().split()))
        ma = map(int, input().split())
        mb = map(int, input().split())
        a = set()
        for c in ma:
            if c <= k:
                a.add(c)
        print(f"a set: {a}")
        b = set()
        # b contains only numbers not in a
        for c in mb:
            if c <= k:
                b.add(c)
        print(f"b set: {b}")

        both_a_and_b = a.union(b)
        print(f"both_a_and_b set: {both_a_and_b}")

        if len(both_a_and_b) < k or len(b) < int(k / 2) or len(a) < int(k / 2):
            print("NO")
            continue

        if len(both_a_and_b) < k:
            print("NO")
            continue

        in_a_not_in_b = a - b  # must get
        print(f"in_a_not_in_b set: {in_a_not_in_b}")
        if len(in_a_not_in_b) > k / 2:
            print("NO")
            continue

        in_b_not_in_a = b - a  # must get
        print(f"in_b_not_in_a set: {in_b_not_in_a}")
        if len(in_b_not_in_a) > k / 2:
            print("NO")
            continue

        # https://www.w3schools.com/python/ref_set_intersection.asp
        in_a_and_in_b = a.intersection(b)
        print(f"in_a_and_in_b set: {in_a_and_in_b}")
        tobe_got_from_a = int(k / 2) - len(in_a_not_in_b)
        print(f"tobe_got_from_a:{tobe_got_from_a}")
        tobe_got_from_b = int(k / 2) - len(in_b_not_in_a)
        print(f"tobe_got_from_b:{tobe_got_from_b}")
        if len(in_a_and_in_b) == tobe_got_from_a + tobe_got_from_b:
            print("YES")
        else:
            print("NO")
```
