```python
from typing import List
from sortedcontainers import SortedList
from queue import PriorityQueue

class Solution:
    def furthestBuilding(self, heights: List[int], bricks: int, ladders: int) -> int:
        used_bricks = PriorityQueue()
        curr_used_bricks = 0
        mub = 0
        
        for index in range(len(heights)-1):
            curr_diff = heights[index+1] - heights[index]
            
            if curr_diff <= 0:
                continue

            if curr_used_bricks + curr_diff <= bricks:
                used_bricks.put(-curr_diff)#log(n)
                curr_used_bricks += curr_diff
                continue

            if ladders == 0:
                return index

            ladders -= 1
            
            # Get max used bricks
            if not used_bricks.empty():
                mub = -used_bricks.queue[0]
                
            if mub > curr_diff:
                curr_used_bricks -= mub
                curr_used_bricks += curr_diff
                used_bricks.get_nowait()
                used_bricks.put(-curr_diff)#log(n)
            
                
        return len(heights) - 1 
                        
```

```python
from typing import List
import heapq 

class Solution:
    def furthestBuilding(self, heights: List[int], bricks: int, ladders: int) -> int:
        curr_used_bricks = 0
        mub = 0
        h = []
        
        for index in range(len(heights)-1):
            curr_diff = heights[index+1] - heights[index]
            
            if curr_diff <= 0:
                continue

            if curr_used_bricks + curr_diff > bricks:
                if ladders == 0:
                    return index

                ladders -= 1
                
                # Get max used bricks
                if h:
                    mub = -heapq.heappop(h)
                    heapq.heappush(h,-mub)
                    
                if mub <= curr_diff:
                    continue
                
                curr_used_bricks -= mub
                heapq.heappop(h)
            
            heapq.heappush(h, -curr_diff)
            curr_used_bricks += curr_diff
                
        return len(heights) - 1 
                        
```
