```python
import cv2
def list_connected_cameras():
    # Get all available cameras
    num_cameras = 10
    available_cameras = []
    for i in range(num_cameras):
        print(f"Checking camera {available_cameras}")
        cap = cv2.VideoCapture(i)
        if not cap.isOpened():
            break
        else:
            available_cameras.append(i)
            ret, frame = cap.read()
            cv2.imshow('Camera', frame)

        cap.release()

    # Print the list of available cameras
    if available_cameras:
        print("Available cameras:")
        for cam in available_cameras:
            print(f"Camera {cam}")
    else:
        print("No cameras detected.")

if __name__ == "__main__":
    list_connected_cameras()
```

```python
import cv2
import subprocess
import re

def list_usb_cameras():
    # Use subprocess to get USB device information
    try:
        output = subprocess.check_output("lsusb", stderr=subprocess.STDOUT)
    except subprocess.CalledProcessError as e:
        print("Error:", e.output.decode())
        return

    # Parse lsusb output to find USB cameras
    usb_cameras = []
    for line in output.decode().split('\n'):
        match = re.search(r'Bus (\d+) Device (\d+): .+ ID (\w+:\w+)', line)
        if match:
            bus, device, device_id = match.groups()
            usb_cameras.append((bus, device, device_id))

    # Print the list of USB cameras
    if usb_cameras:
        print("USB cameras:")
        for idx, cam in enumerate(usb_cameras, start=1):
            print(f"{idx}. Bus {cam[0]}, Device {cam[1]} - {cam[2]}")
    else:
        print("No USB cameras detected.")

if __name__ == "__main__":
    list_usb_cameras()
```
